```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("data.csv")
```

```python
print(df)
```

```python
df.head()
```

```python
print(df.describe)
```

## Visualization

```python
plt.figure(figsize = (5,5))
sns.histplot(df["diagnosis"], bins = 20, kde = True)
plt.title("Cancer")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

## Visualize relationships between variables with scatter plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='diagnosis', y='id', data=df)
plt.title('Relationship between diagnosis and id')
plt.xlabel('diagnosis')
plt.ylabel('id')
plt.show()
```

# Visualize categorical data with count plot


```python
plt.figure(figsize=(10, 6))
sns.countplot(x='diagnosis', data=df)
plt.title('Count of Categorical Values')
plt.xlabel('diagnosis')
plt.ylabel('Count')
plt.show()
```

# Visualize relationships between variables with a heatmap


```python
# Convert 'diagnosis' column to numerical using Label Encoding
label_encoder = LabelEncoder()
df['diagnosis'] = label_encoder.fit_transform(df['diagnosis'])

# Convert any remaining string values to float, ignoring errors
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
plt.figure(figsize=(25,20))
sns.heatmap(df.corr(), annot=True, cmap='cool', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
